```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Result_test_model_8m.csv',sep=';')
data
```

```python
val = {
    'r_mean' : data['r_mean'],
    'r_min'  : data['r_min'],
    'r_max'  : data['r_max'],
    'A_mean' : data['A_mean'],
    'A_min'  : data['A_min'],
    'A_max'  : data['A_max']
}
```

```python

# Assuming val is a dictionary with numeric values
n = 21

# Create lists to store the results
columns = ["Mean", "Standard Deviation", "Standard Error", "Uncertainty"]

data = []

# Append the header row to the data list
data.append(columns)

# Iterate over the keys in val
for i in val.keys():
    mean_value = round(val[i].mean(), 2)
    std_deviation = round(val[i].std(), 2)
    std_error = round((val[i].std() / np.sqrt(len(val[i]))), 2)
    u = round(np.sqrt((sum((val[i] - mean_value)**2)) / (n*(n-1))), 2)
    uncertainty = f"{mean_value} ± {u}"

    # Append the results to the data list
    data.append([mean_value, std_deviation, std_error, uncertainty])

# Transpose the data
transposed_data = list(zip(*data))

# Print the transposed table
for row in transposed_data:
    print("{:<20} {:<13} {:<13} {:<13} {:<13} {:<13} {}".format(*row))
```
