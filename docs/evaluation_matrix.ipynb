```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
x = np.arange(4)
YOLOv8n = [94.40, 65.83, 99.11, 79.11]
YOLOv8s = [95.30, 77.18, 96.44, 85.74]
YOLOv8m = [99.00, 94.30, 99.55, 96.86]
YOLOv8l = [95.50, 85.51, 94.65, 89.85]
YOLOv8x = [97.70, 85.93, 99.33, 92.15]

width = 0.1

fig, ax = plt.subplots(figsize=(12,5))
ax.bar(x-0.2, YOLOv8n, width, edgecolor='k',color='r') 
ax.bar(x-0.1, YOLOv8s, width, edgecolor='k',color='g') 
ax.bar(x,     YOLOv8m, width, edgecolor='k',color='b') 
ax.bar(x+0.1, YOLOv8l, width, edgecolor='k',color='y')
ax.bar(x+0.2, YOLOv8x, width, edgecolor='k',color='c')

ax.yaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_major_formatter('{x:.0f}')
ax.yaxis.set_minor_locator(MultipleLocator(5))
ax.tick_params(right=True, which='both', direction='in')

ax.set_xticks(x, ["mAP", "Precision", "Recall", "F1-score"]) 
ax.set_xlabel("YOLOv8 Models") 
ax.set_ylabel("Scores (%)") 

ax.legend(['YOLOv8n', 'YOLOv8s', 'YOLOv8m', 'YOLOv8l', 'YOLOv8x'], loc='upper center', bbox_to_anchor=(0.5, -0.12),
          fancybox=True, shadow=True, ncol=5)
```
