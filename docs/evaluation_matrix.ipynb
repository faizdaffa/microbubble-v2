```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
mAP = [94.40, 95.30, 99.00, 95.50, 97.70]
P   = [65.83, 77.18, 94.30, 85.51, 85.93]
R   = [99.11, 96.44, 99.55, 94.65, 99.33]
F1  = [79.11, 85.74, 96.86, 89.85, 92.15]
width = 0.2

plt.bar(x-0.3, mAP, width, color='blue') 
plt.bar(x-0.1, P,   width, color='red') 
plt.bar(x+0.1, R,   width, color='yellow') 
plt.bar(x+0.3, F1,  width, color='cyan') 
plt.xticks(x, ['YOLOv8n', 'YOLOv8s', 'YOLOv8m', 'YOLOv8l', 'YOLOv8x']) 
plt.xlabel("YOLO Models") 
plt.ylabel("Scores (%)") 
plt.legend(["mAP", "Precision", "Recall", "F1 Score"], bbox_to_anchor=(1, 1)) 
plt.grid()
plt.show()
```
